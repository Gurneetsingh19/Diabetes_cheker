```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes (1).csv") # Pima Indian Diabtes Database
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.tail()
```

```python
import numpy as np


cols_fix = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']


df[cols_fix] = df[cols_fix].replace(0, np.nan)
```

```python
df.head(10)
```

```python

df['Glucose'].fillna(df['Glucose'].median(), inplace=True)
df['BloodPressure'].fillna(df['BloodPressure'].median(), inplace=True)
df['BMI'].fillna(df['BMI'].median(), inplace=True)


df['Insulin'].fillna(df['Insulin'].median(), inplace=True)
df['SkinThickness'].fillna(df['SkinThickness'].median(), inplace=True)
```

```python
df.head()
```

```python
df.drop("Insulin",axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
update1 = df.drop("Outcome",axis=1)
```

```python
update1
```

```python
df2 = sc.fit(update1)
```

```python
df2 = sc.transform(update1)
```

```python
df2
```

```python
df[["Pregnancies","Glucose","BloodPressure","SkinThickness", "BMI",	"DiabetesPedigreeFunction"	,"Age"]] = df2
```

```python
df.drop("SkinThickness",axis=1,inplace=True)
df
```

```python
df.drop("BloodPressure",axis=1,inplace=True)
df
```

```python
X=df.drop('Outcome',axis=1)
Y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_text,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
Lmodel = lr.fit(x_train,y_train)
```

```python
Lpred = Lmodel.predict(x_text)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,Lpred)
cm
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
print(accuracy_score(y_test,Lpred))
print(precision_score(y_test,Lpred))
print(recall_score(y_test,Lpred))
print(f1_score(y_test,Lpred))
```
